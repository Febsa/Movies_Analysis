```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv("/content/movies.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

**Check Missing Values**

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

**Handling Missing Value**

```python
df.drop(columns = ['id', 'imdb_id', 'homepage', 'cast', 'tagline', 'overview', 'budget_adj'], inplace = True)
```

```python
df.dropna(how = 'any', subset = ['genres', 'director'], inplace = True)
```

```python
df['production_companies'] = df['production_companies'].fillna(0)
df['keywords'] = df['keywords'].fillna(0)
```

```python
df.head()
```

**Insert Columns & Transform Data**

```python
df.insert(3, 'profit', df.revenue - df.budget)
```

```python
df.insert(4, 'roi', df.profit / df.budget)
```

```python
df['popularity'] = df['popularity'].round(2)
df['roi'] = df['roi'].round(2)
```

```python
df.head()
```

```python
df1 = df[['popularity', 'budget', 'revenue', 'profit', 'roi','vote_count', 'vote_average', 'release_year']]
```

**Handling Missing Values**

```python
df1.isnull().sum()
```

```python
df.roi.value_counts()
```

```python
non_finite_values = ~np.isfinite(df['roi'])
```

```python
non_finite_values.sum()
```

```python
df['roi'] = df['roi'].replace([np.inf, -np.inf], np.nan)
```

```python
df1.hist(bins = 20, figsize = (14,12))
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Data")
plt.show()
```

**Top 5 Movies with the biggest Profit**

```python
df10 = df.groupby('original_title')['profit'].sum().sort_values(ascending=False).head(5)

# Check if there is a negative value in the profit
if (df10 < 0).any():
    print("There is a negative value in profit.")
else:
    plt.figure(figsize=(8, 8))
    plt.pie(df10, labels=df10.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title("Top 5 Movies with the biggest profit", color = 'red')
    plt.axis('equal')
    plt.show()
```
